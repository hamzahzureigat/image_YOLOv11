```python

import os
import json
from PIL import Image

# === Configuration ===
DATASET_ROOT = "/Users/admin/Downloads/image"
SPLITS = ["train", "validation", "test"]  # Dataset folder splits

# === Function to Load Dataset ===
def load_dataset(dataset_root):
    data = {}

    for split in SPLITS:
        image_dir = os.path.join(dataset_root, "images", split)
        annot_dir = os.path.join(dataset_root, "annotations", split)

        split_data = []

        # Check if split folders exist
        if not os.path.isdir(image_dir) or not os.path.isdir(annot_dir):
            print(f"Skipping missing split: {split}")
            continue

        for file_name in os.listdir(image_dir):
            if not file_name.endswith(".jpg"):
                continue

            image_path = os.path.join(image_dir, file_name)
            json_filename = file_name + ".json"  # Matches: image.jpg â†’ image.jpg.json
            json_path = os.path.join(annot_dir, json_filename)

            if not os.path.exists(json_path):
                print(f"Warning: Annotation not found for {file_name}")
                continue

            with open(json_path, "r") as f:
                annotation = json.load(f)

            image = Image.open(image_path)

            split_data.append({
                "image_path": image_path,
                "annotation": annotation,
                "image_size": image.size  # (width, height)
            })

        data[split] = split_data

    return data

# === Load the Dataset ===
dataset = load_dataset(DATASET_ROOT)

# === Debug Print ===
for split in SPLITS:
    print(f"{split.capitalize()} samples loaded: {len(dataset.get(split, []))}")

# Print first training sample if available
if dataset.get('train'):
    first = dataset['train'][0]
    print("\nFirst training sample:")
    print("Image path:", first['image_path'])
    print("Image size:", first['image_size'])
    print("Annotation preview:", json.dumps(first['annotation'], indent=2)[:500] + "...")
else:
    print("No training data found.")

```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Function to visualize an image with annotations
def visualize_image_with_annotations(image_data):
    # Open the image
    image = Image.open(image_data['image_path'])
    
    # Prepare the plot
    fig, ax = plt.subplots(1, figsize=(12, 8))
    ax.imshow(image)

    # Loop through objects in the annotation to plot polygons or bounding boxes
    for obj in image_data['annotation']["objects"]:
        # Get the bounding box from polygon points (for YOLO visualization)
        exterior_points = obj["points"]["exterior"]
        if len(exterior_points) > 0:
            # For simplicity, we draw the bounding box using the min/max x/y coordinates
            xmin = min(exterior_points, key=lambda x: x[0])[0]
            xmax = max(exterior_points, key=lambda x: x[0])[0]
            ymin = min(exterior_points, key=lambda x: x[1])[1]
            ymax = max(exterior_points, key=lambda x: x[1])[1]
            
            # Create a rectangle patch for the bounding box
            rect = patches.Rectangle(
                (xmin, ymin), xmax - xmin, ymax - ymin, linewidth=2,
                edgecolor='r', facecolor='none'
            )
            ax.add_patch(rect)

            # You can also display class name (optional)
            ax.text(xmin, ymin, obj['classTitle'], color='white', fontsize=12,
                    bbox=dict(facecolor='red', alpha=0.5, edgecolor='none', boxstyle='round,pad=0.5'))

    plt.axis('off')
    plt.show()

# === Example: Visualize First Training Sample ===
first_sample = dataset['train'][0]
visualize_image_with_annotations(first_sample)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Define a color map for the 13 damage classes only (excluding objects)
class_colors = {
    "rust": 'brown',
    "spalling": 'green',
    "weathering": 'blue',
    "crack": 'orange',
    "efflorescence": 'lightblue',
    "protective equipment": 'purple',
    "cavity": 'pink',
    "hollowareas": 'yellow',
    "drainage": 'red',
    "wetspot": 'lightgreen',
    "graffiti": 'cyan',
    "joint tape": 'magenta',
    "exposed rebars": 'lime'
}

# Function to visualize an image with colored annotations based on damage type (excluding objects)
def visualize_image_with_filtered_annotations(image_data):
    # Open the image
    image = Image.open(image_data['image_path'])
    
    # Prepare the plot
    fig, ax = plt.subplots(1, figsize=(12, 8))
    ax.imshow(image)

    for obj in image_data['annotation']["objects"]:
        class_title = obj['classTitle']

        # Only consider damage types that we care about
        if class_title in class_colors:
            color = class_colors[class_title]

            exterior_points = obj["points"]["exterior"]
            if len(exterior_points) > 2:
                # Draw filled polygon (alpha = transparency)
                polygon = patches.Polygon(
                    exterior_points,
                    closed=True,
                    edgecolor=color,
                    facecolor=color,
                    linewidth=1,
                    alpha=0.5  # 50% transparency
                )
                ax.add_patch(polygon)

                # Optionally label the polygon
                centroid_x = sum(p[0] for p in exterior_points) / len(exterior_points)
                centroid_y = sum(p[1] for p in exterior_points) / len(exterior_points)
                ax.text(centroid_x, centroid_y, class_title, color='black', fontsize=9,
                        bbox=dict(facecolor='white', alpha=0.6, edgecolor='none', boxstyle='round,pad=0.3'))

    plt.axis('off')
    plt.show()

# === Example: Visualize First Training Sample ===
first_sample = dataset['validation'][13]
visualize_image_with_filtered_annotations(first_sample)
```

```python
import os
import json
from PIL import Image

# === Configuration ===
DATASET_ROOT = "/Users/admin/Downloads/image"  # Change if needed
SPLITS = ["train", "validation", "test"]
CLASS_NAMES = [
    "rust", "spalling", "weathering", "crack", "efflorescence",
    "protective equipment", "cavity", "hollowareas", "drainage",
    "wetspot", "graffiti", "joint tape", "exposed rebars"
]
CLASS_ID_MAP = {name: idx for idx, name in enumerate(CLASS_NAMES)}

def convert():
    for split in SPLITS:
        image_dir = os.path.join(DATASET_ROOT, "images", split)
        annot_dir = os.path.join(DATASET_ROOT, "annotations", split)
        label_dir = os.path.join(DATASET_ROOT, "labels", split)
        os.makedirs(label_dir, exist_ok=True)

        for file_name in os.listdir(image_dir):
            if not file_name.endswith(".jpg"):
                continue

            image_path = os.path.join(image_dir, file_name)
            json_path = os.path.join(annot_dir, file_name + ".json")
            label_path = os.path.join(label_dir, file_name.replace(".jpg", ".txt"))

            if not os.path.exists(json_path):
                print(f"Missing annotation for {file_name}")
                continue

            with open(json_path, "r") as f:
                annotation = json.load(f)

            img = Image.open(image_path)
            img_width, img_height = img.size

            yolo_lines = []

            for obj in annotation["objects"]:
                class_title = obj["classTitle"]
                if class_title not in CLASS_ID_MAP:
                    continue  # Skip non-damage objects

                cls_id = CLASS_ID_MAP[class_title]
                points = obj["points"]["exterior"]
                if not points:
                    continue

                # Convert polygon to bounding box
                x_coords = [pt[0] for pt in points]
                y_coords = [pt[1] for pt in points]
                xmin, xmax = min(x_coords), max(x_coords)
                ymin, ymax = min(y_coords), max(y_coords)

                # Normalize to 0-1 for YOLO
                x_center = (xmin + xmax) / 2.0 / img_width
                y_center = (ymin + ymax) / 2.0 / img_height
                box_width = (xmax - xmin) / img_width
                box_height = (ymax - ymin) / img_height

                yolo_lines.append(f"{cls_id} {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}")

            # Write YOLO label file
            with open(label_path, "w") as f:
                f.write("\n".join(yolo_lines))

            print(f"Converted {file_name} â†’ {label_path}")

if __name__ == "__main__":
    convert()
```

```python
import os
import json
import cv2
import numpy as np
from collections import defaultdict
from ultralytics import YOLO
from sklearn.metrics import precision_recall_fscore_support

# === Configuration ===
MODEL_PATH = '/Users/admin/Downloads/image/annotations/validation/runs/detect/bridge_damage_from_scratch/weights/best.pt'
IMAGE_DIR = '/Users/admin/Downloads/image/images/test'
ANNOT_DIR = '/Users/admin/Downloads/image/annotations/test'
IOU_THRESHOLD = 0.5

CLASS_NAMES = [
    "rust", "spalling", "weathering", "crack", "efflorescence",
    "protective equipment", "cavity", "hollowareas", "drainage",
    "wetspot", "graffiti", "joint tape", "exposed rebars"
]
CLASS_ID_MAP = {name: idx for idx, name in enumerate(CLASS_NAMES)}

# === Load model ===
model = YOLO(MODEL_PATH)

def compute_iou(box1, box2):
    xA = max(box1[0], box2[0])
    yA = max(box1[1], box2[1])
    xB = min(box1[2], box2[2])
    yB = min(box1[3], box2[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)
    if interArea == 0:
        return 0.0

    box1Area = (box1[2] - box1[0]) * (box1[3] - box1[1])
    box2Area = (box2[2] - box2[0]) * (box2[3] - box2[1])
    iou = interArea / float(box1Area + box2Area - interArea)
    return iou

# === Main Evaluation ===
y_true = []
y_pred = []

image_files = [f for f in os.listdir(IMAGE_DIR) if f.endswith('.jpg')]

for file_name in image_files:
    image_path = os.path.join(IMAGE_DIR, file_name)
    json_path = os.path.join(ANNOT_DIR, file_name + '.json')

    if not os.path.exists(json_path):
        print(f"[WARNING] Missing annotation for {file_name}, skipping.")
        continue

    img = cv2.imread(image_path)
    img_h, img_w = img.shape[:2]

    # Load GT annotations
    with open(json_path, 'r') as f:
        annot = json.load(f)

    gt_boxes = []
    gt_labels = []

    for obj in annot.get("objects", []):
        cls = obj["classTitle"]
        if cls not in CLASS_ID_MAP:
            continue
        points = obj["points"]["exterior"]
        x_coords = [pt[0] for pt in points]
        y_coords = [pt[1] for pt in points]
        xmin, xmax = min(x_coords), max(x_coords)
        ymin, ymax = min(y_coords), max(y_coords)
        gt_boxes.append([xmin, ymin, xmax, ymax])
        gt_labels.append(CLASS_ID_MAP[cls])

    # Get predictions
    results = model(image_path)
    preds = results[0].boxes.data.cpu().numpy()  # [x1, y1, x2, y2, conf, cls]

    pred_boxes = []
    pred_labels = []
    matched_gt = set()

    for pred in preds:
        x1, y1, x2, y2, conf, cls_id = pred
        pred_box = [x1, y1, x2, y2]
        best_iou = 0
        best_gt_idx = -1
        for i, gt_box in enumerate(gt_boxes):
            iou = compute_iou(pred_box, gt_box)
            if iou > best_iou and i not in matched_gt:
                best_iou = iou
                best_gt_idx = i

        if best_iou >= IOU_THRESHOLD:
            y_true.append(gt_labels[best_gt_idx])
            y_pred.append(int(cls_id))
            matched_gt.add(best_gt_idx)
        else:
            # False positive
            y_true.append(-1)  # no match
            y_pred.append(int(cls_id))

    # Remaining unmatched GTs are false negatives
    for i in range(len(gt_labels)):
        if i not in matched_gt:
            y_true.append(gt_labels[i])
            y_pred.append(-1)  # missed detection

# === Compute Metrics ===
y_true = np.array(y_true)
y_pred = np.array(y_pred)

valid_mask = y_true != -1
valid_preds = y_pred[valid_mask]
valid_trues = y_true[valid_mask]

precision, recall, f1, _ = precision_recall_fscore_support(valid_trues, valid_preds, labels=range(len(CLASS_NAMES)))

# === Report Per Class ===
print("\nðŸ“Š Per-Class Evaluation Results:")
print(f"{'Class':<25} {'Precision':>10} {'Recall':>10} {'F1 Score':>10}")
print("-" * 60)
for idx, name in enumerate(CLASS_NAMES):
    print(f"{name:<25} {precision[idx]:>10.2f} {recall[idx]:>10.2f} {f1[idx]:>10.2f}")
```

```python
import cv2
from ultralytics import YOLO
import os

# === Configuration ===
MODEL_PATH = '/Users/admin/Downloads/image/annotations/test/runs/detect/bridge_damage_from_scratch6/weights/best.pt'
IMAGE_PATH = '/Users/admin/Documents/multi_classifier_data/Corrosion/rust staining/006_qmgnp53n.zni.jpg'  # â† Change this to your test image path
SAVE_OUTPUT = True  # Set to False if you don't want to save output image

CLASS_NAMES = [
    "rust", "spalling", "weathering", "crack", "efflorescence",
    "protective equipment", "cavity", "hollowareas", "drainage",
    "wetspot", "graffiti", "joint tape", "exposed rebars"
]

# === Load Model ===
model = YOLO(MODEL_PATH)

# === Load Image ===
image = cv2.imread(IMAGE_PATH)
if image is None:
    raise FileNotFoundError(f"Image not found at: {IMAGE_PATH}")

# === Run Inference ===
results = model(IMAGE_PATH)
detections = results[0].boxes.data.cpu().numpy()  # [x1, y1, x2, y2, conf, cls_id]

# === Draw Results ===
for det in detections:
    x1, y1, x2, y2, conf, cls_id = det
    cls_id = int(cls_id)
    label = f"{CLASS_NAMES[cls_id]}: {conf:.2f}"

    # Draw box
    cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
    # Put label
    cv2.putText(image, label, (int(x1), int(y1) - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

# === Show and/or Save ===
cv2.imshow("Detected Damage", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

if SAVE_OUTPUT:
    out_path = os.path.join(os.path.dirname(IMAGE_PATH), 'prediction_output.jpg')
    cv2.imwrite(out_path, image)
    print(f"[INFO] Saved annotated image to: {out_path}")
```
